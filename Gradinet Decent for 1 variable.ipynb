```python
import numpy as np
```

```python
def step_gradient(data, learning_rate, m, c):
    m_slope = 0
    c_slope = 0
    N = len(data)
    for i in range(N):
        x = data[i, 0]
        y = data[i, 1]
        m_slope = (-2/N) * (y - m * x - c) * x
        c_slope = (-2/N) * (y - m * x - c)
        new_m = m - learning_rate * m_slope
        new_c = c - learning_rate * c_slope
    return new_m, new_c
```

```python
def gradient_decent(data, learning_rate, num_iteration):
    # initilizing the m, c = 0
    m = 0
    c = 0
    for i in range(len(data)):
        m, c = step_gradient(data, learning_rate, m, c)
        print("cost: ", cost(data, m, c))
    return m, c
```

```python
def cost(data, m, c):
    total_cost = 0
    N = len(data)
    for i in range(N):
        x = data[i, 0]
        y = data[i, 1]
        total_cost += (1/N) * ((y - m * x -c )**2)
    return total_cost
    
```
